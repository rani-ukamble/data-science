```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
pokemon= pd.read_csv('datasets/pokemon.csv') 
```

```python
pokemon
```

```python
#Visualization
```

```python
sns.distplot(pokemon['Attack'])
```

```python
hist_kde=sns.distplot(pokemon['Attack'], kde=False)
```

```python
sns.distplot(pokemon['Attack'],hist=False)
hist_kde.set_title('Attack Capability')
hist_kde.set_xlabel('Attack Point')
hist_kde.set_ylabel('Frequency')
```

```python
sns.distplot(pokemon['Attack'],rug=True)
hist_kde.set_title('Attack Capability')
hist_kde.set_xlabel('Attack Point')
hist_kde.set_ylabel('Frequency')
```

```python
sns.rugplot(pokemon['Attack'],height=0.5,axis='x')
```

```python
sns.distplot(pokemon['Attack'],kde=False, bins=100)
```

```python
sns.distplot(pokemon['Attack'],kde=False, bins=100, vertical=True)
```

#bivariate analysis

```python
sns.jointplot(x='Total',y='Attack',data=pokemon)
```

```python
sns.jointplot(x='Total',y='Attack',data=pokemon ,xlim={300,600},ylim={50,125})
```

```python
sns.jointplot(x='Total',y='Attack',data=pokemon ,kind='hex',color='r')
```

```python
sns.jointplot(x='Total',y='Attack',data=pokemon ,kind='kde',color='r')
```

```python
corr_map=pokemon.corr()
corr_map
```

```python
plt.subplots(figsize=(10,8))
sns.heatmap(corr_map,vmax=0.6,annot=True ,square=False,fmt='.2f',cmap='Reds')
```

```python
sns.lmplot(x='Speed',y='Defense',data=pokemon)
```

```python
sns.lmplot(x='Speed',y='Defense',data=pokemon,col='Type 1')
```

```python
sns.lmplot(x='Speed',y='Defense',data=pokemon,hue='Type 1')
```

```python
sns.jointplot(x='Attack',y='Defense',data=pokemon)
```

```python
sns.pairplot(pokemon,x_vars=['Attack','Speed','Defense'],y_vars=['Total'],kind='reg')
```

#Categorical Analysis ...one of its axis is category

Swarmplot

#swormplot

```python
tips=sns.load_dataset('tips')
tips
```

#Sworm plot

```python
plt.subplots(figsize=(10.,8))
sns.swarmplot(x='day',y='tip',data=tips)
```

```python
plt.subplots(figsize=(10.,8))
sns.swarmplot(x='day',y='tip',data=tips,hue='size')
```

#boxplots

```python
plt.subplots(figsize=(15,8))
sns.boxplot(x='day',y='tip',data=tips)
plt.show()
```

#violin plot

```python
plt.subplots(figsize=(15,8))
sns.violinplot(x='day',y='tip',data=tips)
```

```python
plt.subplots(figsize=(15,8))
sns.violinplot(x='day',y='tip',data=tips,inner=None)
sns.swarmplot(x='day',y='tip',data=tips,color='w')
plt.show()
```

#Bar plots

```python
plt.subplots(figsize=(15,8))
sns.barplot(x='day',y='tip',data=tips,hue='size')
plt.show()
```

#Count plot

```python
plt.subplots(figsize=(15,8))
sns.countplot(x='day',data=tips,palette='Blues_d')
plt.xticks(rotation=90)
```

#Point plot

```python
plt.subplots(figsize=(15,8))
sns.pointplot(x='day',y='tip',data=tips)
plt.show()
```

#Cat plot

```python
sns.catplot(x='day',y='tip',data=tips)
```

```python
sns.catplot(x='day',y='tip',data=tips,kind='bar')
```

```python
sns.catplot(x='day',y='tip',data=tips,kind='violin')
```

```python
sns.catplot(x='day',y='tip',data=tips,kind='swarm')
```

```python
sns.catplot(x='day',y='tip',data=tips,kind='point')
```
